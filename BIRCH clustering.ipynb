```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
X=data.iloc[:,[3,4,]].values
```

```python
from sklearn.cluster import Birch
model = Birch()
model.fit(X) 
```

```python
y=model.predict(X)
y
```

```python
supervised=pd.DataFrame(data)
supervised["Cluster"]=y
supervised
```

```python
supervised.to_csv("Model_Brich.csv",index=False)
```

```python
import seaborn as sns
result=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False)
```
